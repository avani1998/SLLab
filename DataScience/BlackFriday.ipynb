```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('BlackFriday.csv')
```

```python
df.head()
```

```python
df.drop(['Product_ID','Stay_In_Current_City_Years'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.head(20)
```

```python
df['City_Category']=df['City_Category'].fillna('X')
```

```python
df.head()
```

```python
df['City_Category']=df['City_Category'].map({
    'A':'Metro Cities',
    'B':'Small Towns',
    'C':'Villages',
    'X':'Unknown'
})
```

```python
df.head(15)
```

```python
df.rename(columns={'Product_Category_1':'Baseball Caps','Product_Category_2':'Wine tubmlers','Product_Category_3':'Pet Raincoats'},inplace=True)
```

```python
df.head()
```

```python
df['Marital_Status']=df['Marital_Status'].map({
    0:'UnMarried',
    1:'Married'
})
```

```python
df.head(15)
```

```python
interval=(0,8,12,24)
categories=['<8','8-12','>12']
df['ball_cut']=pd.cut(df['Baseball Caps'],interval,labels=categories)
ax=sns.countplot(x='Gender',hue='ball_cut',data=df)
ax.set(title='Try',xlabel='Gender',ylabel='Number')
plt.show()
```

```python
ax=sns.countplot(x='Gender',hue='City_Category',data=df)
ax.set(title='City wise male and female',xlabel='Gender',ylabel='Number')
```
